```python
import re
```

```python
text = "That person wears marvelous trousers."

pattern = 'a' # Any "a" will give a match.
# The findall(pattern, text) returns a list with all the matches
re.findall(pattern, text)
```

```python
text = "That person wears marvelous trousers."

pattern = 'er' # Any "er" will give a match
re.findall(pattern, text)
```

```python
text = "That person wears marvelous trousers."

pattern = '[er]' # Any "e" or "r" will give a match
re.findall(pattern, text)
```

```python
text = "Is it spelled gray or grey?"

pattern = 'gr[ae]y' # Any text containing "gr" followed by "a" or "e" and "y" will give a match
re.findall(pattern, text)
```

```python
text = "This is an A and B conversation - so C your \way out of it."

pattern = '[A-Z]' # Any uppercase letter between A-Z
re.findall(pattern, text)
```

```python
text = "I'm not going to the party because 1) Karen is going, 2) I don't like her, and 3) I already have a headache. 4)-How knows"

pattern = '[1-3]' # Any number character between 1-3 will give a match
print(re.findall(pattern, text))
```

```python
'''
Meta-characters
[]: Match set of characters
.: Match any character except newline (\n)
^: Match characters not listed if within set or match beginning of the string
$: Match end of string
|: Functions as an "OR" operator
'''
print('')
```

```python
text = "My boss asked me to turn in my TPS reports. \nI told her they were done, but they are not."
print(text)
pattern = '.' # Any character except newline. The dot can be searched with "\."
print(re.findall(pattern, text))
```

```python
pattern = '[^a-m]' # Any character excepting the "a-m"
print(re.findall(pattern, text))
```

```python
#pattern = '^My boss'
pattern = '^[Mm]y' # At the beggining of the string any "M" or "m" followed by "y"
print(re.findall(pattern, text))
```

```python
print(text)
pattern = 'they are not.$' # "$" means at the end of the string.
print(re.findall(pattern, text))
```

```python
text = "My boss asked me to turn in my TPS reports. \nI told my boss they were done, but they are not."
#print(text)

pattern = '[Mm]y|TPS|reports' # '[Mm]y' OR 'TPS' OR 'reports'
print(re.findall(pattern, text))
```

```python
'''
specify number of matches:
*: Matches previous character 0 or more times
+: Matches previous character 1 or more times
?: Matches previous character 0 or 1 times (optional)
{}: Matches previous characters however many times specified within:
{n} : Exactly n times
{n,} : At least n times
{n,m} : Between n and m times
'''
print('')
```

```python
text = "The complicit cat interacted with the other cats exactly as we expected."

pattern = "ca*t" # "c" followed by any amount of "a" and "t"
print(re.findall(pattern, text))
```

```python
text = "The complicit cat interacted with the other cccccaaaats exactly as we expected."

pattern = "ca+t" # "c" followed by one or more "a" and "t"
print(re.findall(pattern, text))
```

```python
text = "Is the correct spelling color or colour?"

pattern = "col?our" # ? optional
print(re.findall(pattern, text))
```

```python
text = "Let's see how we can match the following: aw, aww, awww, awwww, awwwww"

pattern = "aw{2}"
print(re.findall(pattern, text))
```

```python
text = "Let's see how we can match the following: aw, aww, awww, awwww, awwwww"

pattern = "aw{2,}"
print(re.findall(pattern, text))
```

```python
text = "Let's see how we can match the following: aw, aww, awww, awwww, awwwww"

pattern = "aw{2,4}"
print(re.findall(pattern, text))
```

```python
'''
Character classes
\w: Any alphanumeric character. equivalent to [A-Za-z0-9_]
\W: Any non-alphanumeric character. equivalent to [^A-Za-z0-9_]
\d: Any numeric character. [0-9]
\D: Any non-numeric character. [^0-9]
\s: Any whitespace characters. [ \t\n\f]
\S: Any non-whitespace characters.
'''
print('')
```

```python
text = "Th1s is going to_be a weird sentence with @ bunch-of-$tuff in it <3."

pattern = '\w'
print(re.findall(pattern, text))
```

```python
text = "Th1s is going to_be a weird sentence with @ bunch-of-$tuff in it <3."

pattern = '\W' 
print(re.findall(pattern, text))
```

```python
text = "Th1s is going to_be a weird sentence with @ bunch-of-$tuff in it <3."
pattern = '\d'
print(re.findall(pattern, text))
```

```python
#extracting words
text = "If you tell the truth, you don't have to remember anything."

pattern = "[\w']+" # Including the apostrophe.
print(re.findall(pattern, text))
```

```python
#words of specified minimum length
pattern = "[\w']{5,}"
print(re.findall(pattern, text))
```

```python
#words of specified maximum length - have to use word boundary \b - need \\b 
# \\b matches the boundary, but not any characters
pattern="\\b[\w']{1,5}\\b"
print(re.findall(pattern, text))
```
