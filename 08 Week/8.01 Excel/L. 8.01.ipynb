```python
import openpyxl
```

```python
from openpyxl import Workbook, load_workbook
```

# Creating a new Excel file

```python
# We create a new "virtual" Excel file. It's virtual because it's defined "in memory"
wb = Workbook()
```

# Adding sheets

```python
# We create two additional "worksheets" in the "workbook"
# When a new sheet is created, the new one becomes the "active sheet"
ws = wb.create_sheet("Mysheet1")
ws = wb.create_sheet("Mysheet2")
```

# Modifying cell's content

We can modifiy the cell content of the first sheet

```python
# This will be introduced in the current "active sheet" "Mysheet2"
ws['A1'] = 42
```

```python
ws.append([1, 2, 3])
```

# Switching to another sheet

```python
ws = wb["Mysheet1"]
```

```python
ws['A1'] = 100
```

```python
ws['A1'].value
```

```python
ws = wb["Mysheet2"]
```

```python
ws['A1'].value
```

# Iterating over a range of cells

```python
for row in regions.iter_rows(min_row=1, max_row=4, min_col=1, max_col=4): # Range(1:4,A:D); index starts in 1.
  print('changing to row...',row)
  for col in row:
    print("Cell: ", col)
    print(col.value)
```
