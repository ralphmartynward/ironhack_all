# Snail and well

A snail falls at the bottom of a 125 cm well. Each day the snail rises 30 cm. But at night, while sleeping, slides 20 cm because the walls are wet. How many days does it take to escape from the well?

````python
# Assign problem data to variables with representative names
# well height, daily advance, night retreat, accumulated distance
well_height = 125
daily_climb = 30
nightly_fall = 20
total_distance = 0

daily_progress = []

# Assign 0 to the variable that represents the solution
total_days = 0

# Assume we begin during the day, we use this boolean to determine if it is day or night
is_day = True

# Write the code that solves the problem
while (total_distance < well_height): # 0 < 125, 30 < 125, 30 - 20 = 10 < 125
    if (is_day == True):
        total_distance = total_distance + daily_climb
        #total_distance += daily_climb
        print("The total distance traveled so far is: ",total_distance)
        #total_dats = total_days + 1
        total_days += 1
        is_day = False
    else:
        total_distance = total_distance - nightly_fall
        #total_distance -= nightly_fall
        print("The total distance traveled so far is: ",total_distance)
        is_day = True
        daily_progress.append(total_distance)
        #total_days += 1
        # Assume we only add days at end of night

# Getting the average        
my_sum = 0
for distance in daily_progress:
    my_sum = my_sum + distance

average = my_sum/len(daily_progress)

import numpy as np
# Getting the standard deviation.
# If ddof is not set, then it will divide by N
# If ddof is set, then it will divide by N - ddof
np.mean(daily_progress)
np.std(daily_progress)  

# Print the result with print('Days =', days)
print('Days =', total_days)
````

````python
# Assign spell power lists to variables

gandalf_power = []

for spell in gandalf:
    gandalf_power.append(POWER[spell])
    
saruman_power = []

for spell in saruman:
    saruman_power.append(POWER[spell])
````

````python
gandalf_power
````

````python
# 2. A sorcerer wins if he succeeds in winning 3 spell clashes in a row.

total_gandalf_consecutive_wins = 0
total_saruman_consecutive_wins = 0

# Execution of spell clashes
for i in range(len(gandalf)):
    #if ( gandalf_power[i] > saruman_power[i] ):
    if (POWER[gandalf[i]] > POWER[saruman[i]]):
        print ("Gandalf wins this round")
        total_gandalf_consecutive_wins += 1
        # Reset loser's wins, since we need to win "3 spell clashes in a row"
        total_saruman_consecutive_wins = 0
        if (total_gandalf_consecutive_wins == 3):
            break
    elif (POWER[gandalf[i]] < POWER[saruman[i]]):
        print ("Saruman wins this round")
        total_saruman_consecutive_wins += 1
        # Reset loser's wins, since we need to win "3 spell clashes in a row"
        total_gandalf_consecutive_wins = 0
        if (total_gandalf_consecutive_wins == 3):
            break
    else:
        # Assume ties do not count against "3 spell clashes in a row" win requirement
        print ("Round ended in a tie")

# check for 3 wins in a row
#Â OR  -> A or B -> A or B must be True to become True
# AND -> A and B -> A and B must be True to become True
if ((total_gandalf_consecutive_wins >= 3) or (total_saruman_consecutive_wins >= 3)):
    print ("We have a winner!")
else:
    print ("No wizard won 3 spell clashes in a row. There are no winners.")
    
# check the winner
if (total_gandalf_consecutive_wins >= 3):
    print ("Gandalf is the winner")
elif (total_saruman_consecutive_wins >= 3):
    print ("Saruman is the winner")
````

# Robin Hood:
We are in a competition to win the archery contest in Sherwood. With our bow and arrows we shoot on a target and try to hit as close as possible to the center.

The center of the target is represented by the values (0, 0) on the coordinate axes.

![](images/arrows.jpg)

## Goals:
* data structures: lists, sets, tuples
* logical operators: if-elif-else
* loop: while/for
* minimum (optional sorting)

## Description:
In the 2-dimensional space, a point can be defined by a pair of values that correspond to the horizontal coordinate (x) and the vertical coordinate (y). The space can be divided into 4 zones (quadrants): Q1, Q2, Q3, Q4. Whose single point of union is the point (0, 0).

If a point is in Q1 both its x coordinate and the y are positive. I leave a link to wikipedia to familiarize yourself with these quadrants.

https://en.wikipedia.org/wiki/Cartesian_coordinate_system

https://en.wikipedia.org/wiki/Euclidean_distance

## Shots
```

          (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2),
          (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]
```

## Tasks
1. Robin Hood is famous for hitting an arrow with another arrow. Did you get it?
2. Calculate how many arrows have fallen in each quadrant.
3. Find the point closest to the center. Calculate its distance to the center. 
4. If the target has a radius of 9, calculate the number of arrows that must be picked up in the forest.

````python
# Variables

points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5),
          (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2),
          (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]
````

````python
import numpy as np
print(np.sort(points))
````

````python
# 1. Robin Hood is famous for hitting an arrow with another arrow. Did you get it?

# Check if any adjacent tuples are equal
for i in range (1, len(points)):
    if (points[i] == points[i-1]):
        print ("We hit an arrow with another arrow!")
        print(points[i])
````

````python
# 2. Calculate how many arrows have fallen in each quadrant.
Q1_total = 0
Q2_total = 0
Q3_total = 0
Q4_total = 0

for x in points:
    if ((x[0] > 0) and (x[1] > 0)):
        Q1_total += 1
    elif ((x[0] < 0) and (x[1] > 0)):
        Q2_total += 1
    elif ((x[0] < 0) and (x[1] < 0)):
        Q3_total += 1
    else:
        Q4_total += 1


print ("Q1 Total = ", Q1_total)
print ("Q2 Total = ", Q2_total)
print ("Q3 Total = ", Q3_total)
print ("Q4 Total = ", Q4_total)    
````

````python
# 3. Find the point closest to the center. Calculate its distance to the center
# Defining a function that calculates the distance to the center can help.

import math
import sys

closest_distance = 100
closest_shots = [] # List to store closest shot to the center.

for x in points:
    distance_from_center = math.hypot(x[0],x[1])
    if (distance_from_center <= closest_distance):
        if (closest_distance == distance_from_center):
            closest_shots.append(x)
        else:
            closest_distance = distance_from_center
            closest_shots = [x]

print("Closest shots are:")
for x in closest_shots:
    print (x)
````

````python
# 4. If the target has a radius of 9, calculate the number of arrows that 
# must be picked up in the forest.

import math

missed_shots = []
radius = 9

for x in points:
    distance_from_center = math.hypot(x[0],x[1])
    if (distance_from_center > radius):
        missed_shots.append(x)
  
print("Missed shots are: ",len(missed_shots))
#for x in missed_shots:
#    print (x)
````
