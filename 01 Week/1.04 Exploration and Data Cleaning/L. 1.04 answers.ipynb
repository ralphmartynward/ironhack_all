## Replacing the missing values

Sometimes, we can replace the missing values by a "reasonable guess". This "guess" can be determined by several techniques:

* Domain knowledge (we already know their source and we know what's the value)

* Mean value of the column

  we can use the Pandas method .mean() to compute the mean of a column

* Median value of the column

  we can use the Pandas method .median() to compute the median of a column

* Other more sophisticated approaches

To fill the missing values of a column, we can use the .fillna(value) method. This method, will replace **all** the missing values with the *value* provided within parentheses.

```python
# strategy: impute a value for the missing value (fill in a value that we choose or calculate)

import numpy as np

mean_median_home_value = donors['median_home_val'].mean()
mean_median_home_value
donors['median_home_val'] = donors['median_home_val'].fillna(mean_median_home_value)
```

```python
def clean_gender(x):
    if x in ['M', 'Male', 'male']: # Male gender
        return 'Male'
    elif x[0] in ['F','f']: # Female gender
        return 'Female'
    else: # Other
        return 'U'

def clean_state(x):
    x = x.upper()
    x = x.strip()
    if x in ['AZ', 'ARIZONA']:
        return 'AZ'
    elif x in ['CA', 'CALIFORNIA', 'CALI']:
        return 'CA'
    elif x in ['TN', 'TENNESSEE']:
        return 'TN'
    else:
        return x
```

```python
def clean_dataset(df):
    df2 = df.copy()
    #df2['gender'] = df2['gender'].apply(clean_gender)
    #df2['state'] = df2['state'].apply(clean_state)
    df2['gender'] = list(map(clean_gender,df2['gender']))
    df2['state'] =  list(map(clean_state,df2['state']))
    
    return df2
```
