# Advanced automated filtering methods: Variance Threshold

One way to filter out features/columns is to look at how different are the values in a given column of a dataframe. This concept is given by the **variance** or the **standard deviation**. 

The idea is quite simple: if the values of a column are the same, this column is not giving us information to discriminate between different rows of our dataframe. Therefore, this column is useless, why we should keep it?

However, there are several caveats here:

* Having all the values equal is will be an extreme case and this will almost never happen. Therefore, we need to think in stablishing a threshold on the variance.

* All the columns have to be in the same scale to use this method because otherwise, the columns with the biggest range, will likely have higher variance.

```python
# to look at VarianceThresholds we need all the variable to be on the same scale
# we will use MinMaxScaler for this
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(numerical)
numerical_scaled = scaler.transform(numerical)
```

```python
numerical_scaled_df = pd.DataFrame(numerical_scaled, columns = numerical.columns)
numerical_scaled_df.head()
```

Fortunatelly enough, Sklearn has a function to do this feature selection by variance for us. We only need to provide a threshold for the variance. If the variance of a column is **lower than the threshold** the column will be droped.

```python
from sklearn.feature_selection import VarianceThreshold

var_threshold = 0.02
# sel = VarianceThreshold() has a default value for variance = 0.00
sel = VarianceThreshold(threshold=(var_threshold))

# 1) default is threshold = 0 :eliminate all columns that are identical for all rows
# 2) in practice we would scale the columns first, and then apply threshold, or apply 
#    a different threshold for different columns

# This determines which columns that have a variance less than this threshold
sel = sel.fit(numerical_scaled_df)
temp = sel.transform(numerical_scaled_df)
temp_df = pd.DataFrame(temp)
print(numerical_scaled_df.shape)
print(temp_df.shape)
```

```python
# We can know what were the variances of all the columns.
sel.variances_
```

We can also know what are the columns that are kept.

```python
sel.get_feature_names_out()
```

```python
#This gives you the result as True and False for the columns that we
# selected and those which were removed, respectively.
#sel.variances_ > var_threshold
sel.get_support()
var_list = list(sel.get_support())
var_list
#list(zip(list(numerical.columns), var_list))
```

```python
# We can use this to know if a column was kept or not.
list(zip(numerical.columns, var_list))
```

```python
# As we have a list of tuples, we can know what are the names of the columns that were ketp using a list comprehension.
cols_kept = [tupple[0] for tupple in list(zip(numerical.columns, var_list)) if tupple[1] == True]
# [value1 if condition == True else value2 for elem in list]
# [value1 for elem in list if condition == True]
# [value1 for elem in list]
cols_kept
```

# Recursive Feature Elimination

The Recursive Feature Elimination method.selection needs as an input a regression or classification model and returns the best features for the given model. Therefore **is tied to a model** and the selected features will be the best for a given model. If we change the model, we will get other features.

Let's try here to use this method using a linnear regression model.

The RFE method will do is to train as all linnear models with (n < number_of_columns) features and return the n features of the model with the best R2.

```python
#C(79, 20) = 79!/(20!(79-20)!)
```

```python
from sklearn import linear_model
from sklearn.feature_selection import RFE

lm = linear_model.LinearRegression()
rfe = RFE(lm, n_features_to_select=20, verbose=True)
rfe.fit(X, y)
```

```python
# After we run the algorithm, it labels the top features as 1 and the rest are marked in an increasing order of importance.
df = pd.DataFrame(data = rfe.ranking_, columns=['Rank'])
#pd.DataFrame(data = rfe.ranking_, columns=['Rank'])
df['Column_name'] = pd.DataFrame(numerical).columns
#df.head()
df[df['Rank']==1] # The selected features will have Rank = 1
```

What about the multicollinearity of the select features? This method doesn't take into account this.

```python
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(X[df[df['Rank']==1]['Column_name'].values.tolist()].corr(),annot=True, ax = ax)
plt.show()
```
