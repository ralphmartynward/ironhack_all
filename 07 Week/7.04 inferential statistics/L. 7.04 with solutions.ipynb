```python
#Sample 30 passengers from 3rd class
# random_state=13, random.seed()
c3_sample = data[data['Pclass']==3]['Fare'].sample(30, random_state=1)
c3_sample.head(30)
```

```python
#display(data[data['Pclass']==3]['Fare'].mean())
display(c3_sample.mean()) 
# your value for the mean may be diffrent from the teacher's, and this may affect the p-value
```

```python
#Hypothesis: H0: mean fare in 3rd class was 17
#            H1: mean fare in 3rd class != 17
#compute statistics and p-value
st.ttest_1samp(c3_sample, 17)
```

```python
#for the single tailed experiment
# H0: mean fare in 3rd class was >= 17
# H1: mean fare in 3rd class < 17
print('p value (single tailed): ',st.ttest_1samp(c3_sample,17, alternative="less")[1]/2)
```

Note different samples may produce different results, leading to a different decision on the hypothesis.

# Matched pairs

```python
#Example: Matched Pair.
#The data in the two samples is dependent. In the data we are studying the blood pressure before and after the treatment.
```

```python
blood_pressure = pd.read_csv('/content/drive/MyDrive/CURR-v3.X-JAN2023/UNIT7/DAY2/7.04 inferential statistics/blood_pressure.csv')
blood_pressure.head()
```

```python
st.ttest_rel(blood_pressure['after'], blood_pressure['before'])
```

$$H0: \mu_{1} = \mu_{2}; \mu_{1}-\mu{2}=0$$
$$H1: \mu_{1} \neq \mu_{2}; \mu_{1}-\mu{2}\neq 0 $$

```python
# We reject the null hypothesis. p_value <<< 0.05
```

```python
# P_value = P(S|H0=True), H0: No effect -> mean_blood_preasure_after = mean_blood_preasure_before
```

```python
#negative statistic means \mu_after-\mu_before < 0   ---->  \mu_after < \mu_before 
```

# Pooled samples / independent samples

```python
#Independent Samples
#For two groups where we cannot match the observations to one another.
#In this case transactions from a website with different interfaces (a, b)
```

# ANOVA

```python
#ANOVA test
#to compare the means of multiple groups at the same time
#why not just multiple t-tests?
```

```python
interest_r = pd.read_csv('/content/drive/MyDrive/CURR-v3.X-JAN2023/UNIT7/DAY2/7.04 inferential statistics/rate_by_city.csv')
interest_r.head(60)
```

```python
interest_r.groupby(['City']).agg({"Rate":np.mean})
```

```python
interest_r['city_count'] = interest_r.groupby('City').cumcount() ##is the new index 
#interest_r
interest_r_pivot = interest_r.pivot(index='city_count', columns='City', values='Rate')
#interest_r_pivot
interest_r_pivot.columns = ['City_'+str(x) for x in interest_r_pivot.columns.values]
interest_r_pivot.head(10)
```

```python
interest_r_pivot.mean()
```

```python
# H0 for ANOVA is always that the means of the various groups are the same
# H1 is that they are not the same
st.f_oneway(interest_r_pivot.City_1,interest_r_pivot.City_2,interest_r_pivot.City_3,interest_r_pivot.City_4,interest_r_pivot.City_5,interest_r_pivot.City_6)
```
